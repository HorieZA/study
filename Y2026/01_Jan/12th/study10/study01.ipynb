### 비교 연산자

```python
print(1 == 1)
print(1 != 1)
print(1 < 2)
print(1 > 1)
print(1 <= 2)
print(1 >= 2)
```

### 논리 연산자

```python
print(1 == 1 and 1 != 2)
print(1 == 1 or 1 != 1)
print(not 1 != 1)
print(not 1 == 1)
```

```python
print(1 == 1 and 1 != 2 or 1 != 1)
print(not 1 == 1 and 1 != 2 or 1 != 1)
print(not 1 == 1 and (1 != 2 or 1 != 1))
```

### 구구단 출력하기

```python
for x in range(2,4) :
  for y in range(1,10) :
    print(f"{x}*{y} = {(x*y)}")
```

```python
# 파스칼 출력
for i in range(5,(i+1)) :
    print = ""
    num = 1
    for s in range((5-i-1),(s+1)) :
        print+=" "
        for j in range(i,(j+1)) :
            print+=num
            num=num*(i-j)/(j+1)
    print(print)
```

```python
import math
a = ""
for i in range(0, 2*3-1, i+1) :
    a=""
    for j in range(0, 2*3-1, j+1) :
        if(math.abs(i-4)+math.abs(j-4) <= 4) : a+="◆"
        else : a+=" "
    print(a)
```
